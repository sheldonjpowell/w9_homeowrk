```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('./AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighbourhood groups / which shows up the most
nei_group = air_bnb.sort_values('neighbourhood_group', ascending = False)
nei_group.head(1).reset_index(drop=True)
```

```python
def many(group):
#     si = 0
#     group_set = {}
    group_s = []
    for g in group:
        group_s.append(g)
    count = pd.value_counts(np.array(group_s))
    print(count)
    
many(nei_group['neighbourhood_group'])    
        
```

```python
# Are private rooms the most popular in Manhattan?
prooms = air_bnb.sort_values('room_type', ascending = False)
prooms.head(1).reset_index(drop=True)
```

```python
def rooms(typeof):
#     si = 0
#     group_set = {}
    type_rooms = []
    for type in typeof:
        type_rooms.append(type)
    count = pd.value_counts(np.array(type_rooms))
    print(count)
    
rooms(prooms['room_type'])
```
